```python
from pywdf.examples.sallenkeyfilter import SallenKeyFilter

SAMPLERATE = 44100
cutoff = 1000 # Hz
q_factor = 1
sk_filter = SallenKeyFilter(SAMPLERATE, cutoff, q_factor)
```

```python
cutoff = 100 # Hz
sk_filter.set_params(cutoff, q_factor)
sk_filter.plot_freqz()
```

```python
from pywdf.examples.diodeclipper import DiodeClipper

cutoff = 1000 # Hz
input_gain = 0 # dB
output_gain = 0 # dB
num_diodes = 2
dc = DiodeClipper(SAMPLERATE, cutoff, input_gain, output_gain, num_diodes)
```

```python
dc.plot_freqz()
```

```python
cutoff = 100
dc.set_cutoff(cutoff)
dc.plot_freqz()
```

```python
dc.AC_transient_analysis()
```
